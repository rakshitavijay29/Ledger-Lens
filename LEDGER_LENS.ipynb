```python
df[[
'transaction_amount',
'amount_deviation',
'txn_balance_ratio',
'desc_frequency',
'balance_change'
]].isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.ensemble import IsolationForest
iso=IsolationForest(
    n_estimators=100,
    contamination=0.02,
    random_state=42
)
iso.fit(X_scaled)
```

```python
df['anomaly']=iso.predict(X_scaled)
```

```python
df['anomaly'].value_counts()
```

```python
anomalies=df[df['anomaly']==-1]
anomalies[['ac_no','date','transaction_amount',
'amount_deviation',
'desc_frequency',
'balance_change']].head(20)
```

```python
anomalies['ac_no'].value_counts()
```

```python
acc_stats=df.groupby('ac_no').size()
anomaly_stats=anomalies.groupby('ac_no').size()
```

```python
risk_score = (
    anomaly_stats / acc_stats * 100
).sort_values(ascending=False)

```

```python
risk_score.reset_index(name='risk_percentage')
```

```python
import matplotlib.pyplot as plt

risk_score.plot(kind='bar',figsize=(10,5))
plt.ylabel('anomaly_per')
plt.title('Risk Ranking of Accounts')
plt.show()
```

```python
df['anomaly_score'] = iso.decision_function(X_scaled)
```

```python
fraud_report=anomalies.groupby('ac_no').agg(
    total_anomaly=('anomaly','count'),
    largest_trans=('transaction_amount','max'),
    avg_trans=('transaction_amount','mean'),
     first_suspicious_date=('date', 'min'),
    last_suspicious_date=('date', 'max'))
```

```python
fraud_report.head(15)
```

```python
total_transactions = df.groupby('ac_no').size()
```

```python
fraud_report['total_transactions'] = total_transactions
```

```python
fraud_report['anomaly_percentage'] = (
    fraud_report['total_anomaly']
    / fraud_report['total_transactions']
) * 100
```

```python
def risk_level(x):

    if x >= 10:
        return "High"

    elif x >= 5:
        return "Medium"

    else:
        return "Low"
```

```python
fraud_report['risk_level'] = fraud_report['anomaly_percentage'].apply(risk_level)
fraud_report['risk_level'].head(11)
```
